```python
# dding inf parameters
from google.colab import drive
drive.mount('/content/drive')

# Install the antiglitch package from your local directory
!pip install /content/drive/MyDrive/Colab Notebooks/Project Notebooks/one tomte

from antiglitch.utils import Snippet
import matplotlib.pyplot as plt

# Adjust the data directory to point to your local directory on Google Drive
from antiglitch import utils
utils.datadir = '/content/drive/MyDrive/Colab Notebooks/Project Notebooks/one koi fish/'


import numpy as np
from numpy.random import default_rng

# Create an empty list to store the generated snippets
snippets = []

rng = default_rng(seed=7678735)
# Loop to create 1000 instances of glitches with noise
for _ in range(1000):
    # Creating a new Snippet instance
    snip = Snippet('L1', 'koi', 0)

    # Setting up inference parameters
    inf = {'amp_r': np.random.uniform(500, 900),    # Amplitude real part
           'amp_i': 20.,       # Amplitude imaginary part
           'f0': np.random.uniform(20, 30),         # Central frequency
           'gbw': np.random.uniform(0.2, 2.0),      # Bandwidth
           'time': 0.}

    # Setting the inference parameters
    snip.set_infer(inf)

    # Generating random noise
    noise = rng.normal(size=len(snip.whts))

    # Adding noise to the glitch
    glitch_with_noise = snip.glitch + noise

    # Storing the snippet and glitch with noise
    snippets.append((snip, glitch_with_noise))
```

```python
plt.plot(snip.glitch)
```

```python
# Plotting the first 10 instances of glitch with noise
for i, (snip, glitch_with_noise) in enumerate(snippets[:10]):
    plt.figure()
    plt.plot(snip.glitch)
    #plt.plot(glitch_with_noise)
    plt.title(f'Snippet {i+1}')
    plt.legend(['Original Glitch', 'Glitch with Noise'])
    plt.show()
```

DOWNLOAD


```python
import os
import numpy as np
from scipy.signal import welch

# Create a directory to store the .npz files if it doesn't exist
directory = '/content/drive/MyDrive/Colab Notebooks/Project Notebooks/MODELS /KOI1000'
if not os.path.exists(directory):
    os.makedirs(directory)

# Save each glitch with noise and its corresponding PSD as .npz and .npy files
for i, (snip, glitch_with_noise) in enumerate(snippets):
    # Calculate the PSD using Welch's method
    f, Pxx = welch(glitch_with_noise, fs=1.0)

    # Save glitch with noise as .npz file
    glitch_filename = f'koifish-{i+1}.npz'
    glitch_filepath = os.path.join(directory, glitch_filename)
    np.savez(glitch_filepath, glitch_with_noise=glitch_with_noise)

    # Save PSD as .npy file
    psd_filename = f'psd_{i+1}.npy'
    psd_filepath = os.path.join(directory, psd_filename)
    np.save(psd_filepath, Pxx)
```
